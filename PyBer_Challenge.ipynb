### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
#urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
#suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
#rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

#urban_ride_count = sum(urban_cities_df.groupby(["city"]).count()["ride_id"])
#suburban_ride_count = sum(suburban_cities_df.groupby(["city"]).count()["ride_id"])
#rural_ride_count = sum(rural_cities_df.groupby(["city"]).count()["ride_id"])
#print(suburban_ride_count)

total_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
print(total_ride_count)
```

```python
# 2. Get the total drivers for each city type
#urban_driver_count = urban_cities_df.groupby(["city"]).sum()["driver_count"]
#suburban_driver_count = suburban_cities_df.groupby(["city"]).sum()["driver_count"]
#rural_driver_count = rural_cities_df.groupby(["city"]).sum()["driver_count"]
#print(urban_driver_count)

total_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
print(total_driver_count)
```

```python
#  3. Get the total amount of fares for each city type
#urban_total_fare = sum(urban_cities_df.groupby(["city"]).sum()["fare"])
#suburban_total_fare = sum(suburban_cities_df.groupby(["city"]).sum()["fare"])
#rural_total_fare = sum(rural_cities_df.groupby(["city"]).sum()["fare"])
#print(suburban_total_fare)

total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
print(total_fare)
```

```python
#  4. Get the average fare per ride for each city type. 
#urban_avg_fare_per_ride = urban_total_fare / urban_ride_count
#suburban_avg_fare_per_ride = suburban_total_fare / suburban_ride_count
#rural_avg_fare_per_ride = rural_total_fare / rural_ride_count
#print(urban_avg_fare_per_ride)

avg_fare_per_ride = total_fare / total_ride_count
print(avg_fare_per_ride)
```

```python
# 5. Get the average fare per driver for each city type. 
#urban_avg_fare_per_driver = urban_total_fare / urban_driver_count
#suburban_avg_fare_per_driver = suburban_total_fare / suburban_driver_count
#rural_avg_fare_per_driver = rural_total_fare / rural_driver_count

avg_fare_per_driver = total_fare / total_driver_count
print(avg_fare_per_driver)
```

```python
#  6. Create a PyBer summary DataFrame. 
#total_rides = [urban_ride_count, suburban_ride_count, rural_ride_count]
#total_drivers = [urban_driver_count, suburban_driver_count, rural_driver_count]
#total_fares = [urban_total_fare, suburban_total_fare, rural_total_fare]
#avg_fare_per_ride = [urban_avg_fare_per_ride, suburban_avg_fare_per_ride, rural_avg_fare_per_ride]
#avg_fare_per_driver = [urban_avg_fare_per_driver, suburban_avg_fare_per_driver, rural_avg_fare_per_driver]

#   [{"Total Rides": total_ride_count,      
#     "Total Drivers": total_driver_count,      
#     "Total Fares": total_fares,      
#     "Average Fare per Ride": avg_fare_per_ride,       
#     "Average Fare per Driver": avg_fare_per_driver}])

#pyber_ride_driver = pd.merge(total_ride_count, total_driver_count, how="left", on=["type"])
#pyber_rd_fare = pd.merge(pyber_ride_driver, total_fares, how="left", on=["type"])
#pyber_rd_fare

pyber_summary_df = pd.DataFrame()
pyber_summary_df["Total Rides"] = total_ride_count
pyber_summary_df["Total Drivers"] = total_driver_count
pyber_summary_df["Total Fares"] = total_fare
pyber_summary_df["Average Fare per Ride"] = avg_fare_per_ride
pyber_summary_df["Average Fare per Driver"] = avg_fare_per_driver

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df.style.format({"Total Fares": "${:.2f}", "Average Fare per Ride": "${:.2f}", "Average Fare per Driver": "${:.2f}"})
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_type_date_df = pyber_data_df.groupby(["date", "type"]).sum()["fare"]
pyber_type_date_df.head(10)
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_type_date_df = pyber_type_date_df.reset_index()
pyber_type_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_td_pivot_df = pyber_type_date_df.pivot(index = "date", columns = "type", values = "fare")
pyber_td_pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
mask = (pyber_td_pivot_df.index > '2019-01-01') & (pyber_td_pivot_df.index <= '2019-04-29')
pyber_td_JanMar19_df = pyber_td_pivot_df.loc[mask]
pyber_td_JanMar19_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pyber_td_JanMar19_df.index = pd.to_datetime(pyber_td_JanMar19_df.index)
pyber_td_JanMar19_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_td_JanMar19_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resample_pyber_week_df = pyber_td_JanMar19_df.resample("W").sum()
resample_pyber_week_df
```

```python
# 9. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
fig, ax = plt.subplots(figsize = (30, 10))
ax.plot(resample_pyber_week_df)
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.legend(resample_pyber_week_df.columns)

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
